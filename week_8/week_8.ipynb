First we are going to import our data modules.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Next we want import the data and begin replacing some NaN values.

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data', sep='\s+', header=None)
data

data.columns=['mpg', 'cylinders', 'displacement', 'HP', 'weight', 'acceleration','model year', 'origin', 'car name']
data.replace('?', np.NaN)
data.HP = pd.to_numeric(data.HP, errors = 'coerce')
```

Now if we tail this data set we cab see that the columns names are replaced and everything is in tact.

```python
data.tail()
```

Next we want to replace the origins with the correct. We will map out the names and insert them into the data set.

```python
origin = {1:'USA', 2:'Asia',3:'Europe'}

data.origin = data.origin.map(origin)
data.head()
```

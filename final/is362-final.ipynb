Let make our imports like usual.

```python
import pandas as pd
import seaborn as sns
import math as pi
import matplotlib.pyplot as plt
%matplotlib inline
```

Now lets get our data from our repo. We are going to head the data to see what the CSV contains for columns. 

```python
data = pd.read_csv('https://raw.githubusercontent.com/kneeks/IS-362/master/final/Pokemon.csv', 
                   sep=',')
data.head()
```

Now we are going to drop some columns are sort by HP. 

```python
data.sort_values(['HP'], ascending = False).head(1)
data = data.drop(['#', 'Total', 'Type 1', 'Type 2', 'Generation', 'Legendary'],1)
data
```

Now that we have our data we can begin graphing.

```python
sns.set_style("whitegrid")
sns.jointplot(x="HP", y="Defense", data=data);
```

The above graph demonstrate the defense points that a pokemon has per HP.

Below is a statistical boxdiagram of all the data and each type of attribute.

```python
sns.boxplot(data=data)
plt.show()
```

To conlcude this data HP is based on the defense of the pokemon. Another observation shows that there are a handful of strong pokemon according to the box diagram above.
